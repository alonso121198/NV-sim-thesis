# Cargamos data con y sin corriente electrica

```python
import numpy as np
import matplotlib.pyplot as plt

campos_con_corriente = np.load('./Resultados-tiettene-con-corriente/campos-magneticos-calculadas-64pixeles-en-1.npy')
campos_sin_corriente = np.load('./Resultados-tiettene-sin-corriente/campos-magneticos-calculadas-64pixeles-en-1.npy')

campos_magnetico = (campos_con_corriente - campos_sin_corriente)
```

# Grafico de data con y sin corriente electrica

## Sin corriente

```python
# Crear una figura con 6 subplots organizados en 2 filas y 3 columnas
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

# Iterar sobre los subplots y las imágenes
for ax, idx in zip(axes.flat, range(0, 4)):
    campo = campos_sin_corriente[:,:,idx]
    im = ax.imshow(campo, cmap='bwr')
    ax.set_title(f'Imagen {idx}')
    ax.axis('off')  # Opcional: desactivar los ejes
    
    # Añadir una barra de color para cada imagen
    cbar = plt.colorbar(im, ax=ax, orientation='vertical')
    cbar.ax.set_ylabel('Intensidad', rotation=270, labelpad=10)
    
# fig.savefig('./campo-magnetico-sin-corriente-calculadas-64pixeles-en-1.png')
plt.show()
```

## Con corriente

```python
# Crear una figura con 6 subplots organizados en 2 filas y 3 columnas
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

# Iterar sobre los subplots y las imágenes
for ax, idx in zip(axes.flat, range(0, 4)):
    campo = campos_con_corriente[:,:,idx]
    im = ax.imshow(campo, cmap='bwr')
    ax.set_title(f'Imagen {idx}')
    ax.axis('off')  # Opcional: desactivar los ejes
    
    # Añadir una barra de color para cada imagen
    cbar = plt.colorbar(im, ax=ax, orientation='vertical')
    cbar.ax.set_ylabel('Intensidad', rotation=270, labelpad=10)
    
# fig.savefig('./campo-magnetico-con-corriente-calculadas-64pixeles-en-1.png')
plt.show()
```

# Calculo de campo magnetico total

```python
for i in range(campos_magnetico.shape[0]):
    for j in range(campos_magnetico.shape[1]): 
        
        B = (campos_magnetico[i,j,0]**2 + campos_magnetico[i,j,1]**2 + campos_magnetico[i,j,2]**2)**(1/2)
        campos_magnetico[i,j,3] = B
```

```python
# Crear una figura con 6 subplots organizados en 2 filas y 3 columnas
fig, axes = plt.subplots(2, 2, figsize=(15, 15))

# Iterar sobre los subplots y las imágenes
for ax, idx in zip(axes.flat, range(0, 4)):
    campo = campos_magnetico[:,:,idx]
    im = ax.imshow(campo, cmap='bwr')
    ax.set_title(f'Imagen {idx}')
    ax.axis('off')  # Opcional: desactivar los ejes
    
    # Añadir una barra de color para cada imagen
    cbar = plt.colorbar(im, ax=ax, orientation='vertical')
    cbar.ax.set_ylabel('Intensidad', rotation=270, labelpad=10)

fig.savefig('./campo-magnetico-total-calculadas-64pixeles-en-1.png')
plt.show()
```
