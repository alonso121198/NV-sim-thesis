# Cargamos data con y sin corriente electrica

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

import scienceplots # Para obtener graficos de alta calidad
plt.style.use(['science','notebook','grid','std-colors'])
#plt.style.use(['nature','no-latex'])

campos_con_corriente = np.load('./Con_corriente_guardados/Datos_obtenidos/CamposBcal.npy')
campos_sin_corriente = np.load('./Sin_corriente_guardados/Datos_obtenidos/CamposBcal.npy')

campos_magnetico = (campos_con_corriente - campos_sin_corriente)
```

# Grafico de data con y sin corriente electrica

## Sin corriente

```python
# esto para escribir los titulos de imagenes
def switch(case):
    if case == 0:
        return "$B_x (\mathrm{\mu T})$"
    elif case == 1:
        return "$B_y (\mathrm{\mu T})$"
    elif case == 2:
        return "$B_z (\mathrm{\mu T})$"
    elif case == 3:
        return "$|| \mathbf{B}_0 (\mathrm{\mu T})|| $"

# Crear una figura con 6 subplots organizados en 2 filas y 3 columnas
fig, axes = plt.subplots(2, 2, figsize=(18, 18))

# Iterar sobre los subplots y las imágenes
colores = ["RdPu","Purples","Purples","Greys"]
for ax, idx,color in zip(axes.flat, range(0, 4),colores):
    campo = campos_sin_corriente[:,:,idx]
    im = ax.imshow(campo, cmap=color)
    
    comp_camp = switch(idx)
    ax.set_title(fr'{comp_camp}',fontsize=30)
    ax.axis('off')  # Opcional: desactivar los ejes
    
    # Añadir una barra de color para cada imagen
    cbar = plt.colorbar(im, ax=ax, orientation='vertical',fraction=0.045)
    cbar.ax.tick_params(labelsize=16)
    cbar.ax.set_ylabel('Intensidad', rotation=270, labelpad=25,fontsize=25)
    cbar.formatter = ticker.FormatStrFormatter('%.0f')  # Para 2 decimales

plt.subplots_adjust(wspace=0.3)
    
fig.savefig('./campo-magnetico-sin-corriente-resultado.png')
plt.show()
```

## Con corriente

# Calculo de campo magnetico total

```python
for i in range(campos_magnetico.shape[0]):
    for j in range(campos_magnetico.shape[1]): 
        
        B = (campos_magnetico[i,j,0]**2 + campos_magnetico[i,j,1]**2 + campos_magnetico[i,j,2]**2)**(1/2)
        campos_magnetico[i,j,3] = B
```

```python
# esto para escribir los titulos de imagenes
def switch(case):
    if case == 0:
        return "$\delta B_x (\mathrm{\mu T})$"
    elif case == 1:
        return "$\delta B_y (\mathrm{\mu T})$"
    elif case == 2:
        return "$\delta B_z (\mathrm{\mu T})$"
    elif case == 3:
        return "$|| \mathbf{\delta B}_I (\mathrm{\mu T})|| $"

# Crear una figura con 6 subplots organizados en 2 filas y 3 columnas
fig, axes = plt.subplots(2, 2, figsize=(18, 18))

# Iterar sobre los subplots y las imágenes
colores = ["bwr","bwr","bwr","Greys"]
for ax, idx,color in zip(axes.flat, range(0, 4),colores):
    campo = campos_magnetico[:,:,idx]
    if idx < 3:
        im = ax.imshow(campo, cmap=color,vmin=-100, vmax=100)
    else: 
        im = ax.imshow(campo, cmap=color)
    
    comp_camp = switch(idx)
    ax.set_title(fr'{comp_camp}',fontsize=30)
    ax.axis('off')  # Opcional: desactivar los ejes
    
    # Añadir una barra de color para cada imagen
    cbar = plt.colorbar(im, ax=ax, orientation='vertical',fraction=0.045)
    cbar.ax.tick_params(labelsize=16)
    cbar.ax.set_ylabel('Intensidad', rotation=270, labelpad=25,fontsize=25)
    cbar.formatter = ticker.FormatStrFormatter('%.0f')  # Para 2 decimales

plt.subplots_adjust(wspace=0.25)

fig.savefig('./CampoMagdiferencialTettiene.png')
plt.show()
```
