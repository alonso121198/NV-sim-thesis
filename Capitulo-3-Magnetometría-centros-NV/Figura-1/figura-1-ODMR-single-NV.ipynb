## ODMR de un unico centro NV

El perfil puede ser modelado por dos lorentzianas
$$
\frac{I(\nu)}{I(0)} = 1 - f(\nu) \qquad f(\nu) = C_1\left( \frac{ \Gamma^2}{(\nu-\nu_{+})^2 + \Gamma^2} \right) + C_2\left(\frac{ \Gamma^2 }{ (\nu-\nu_{-})^2 + \Gamma^2}  \right) \qquad \nu_{\pm} = D \pm \gamma B_z
$$

donde:
- $\frac{I(\nu)}{I(0)}$ respresenta la disminución inducida por la normalizada intensidad (Fluorescencia)
- $\Gamma$ corresponde al ''ancho mitad a la mitad del máximo'' (Half-width at half-maxmium ''HWHM'') de cada lorentziana
- $C$ detremina el ''contraste'', la caida fraccional en intensidad cuando el microondas esta sintonizado en resonancia
- $\nu_{+}$ y $\nu_{-}$ las frecuencias de transición 
- $\gamma = 28 GHz/T $
- $D = 2.87 GHz$

```python
# Parametros a usar 
nu_menos_prueba = 2868
nu_mas_prueba = 2875.5
C1 = 0.2
C2 = 0.2
HWHM = 0.05

# Definimos nu_{+} y nu_{-}
def nu_menos(B_z):
    return 2.87-B_z*28
def nu_mas(B_z):
    return 2.87+B_z*28

def func(fmw,B_z):
	return 1 - C1*( (HWHM**2)/(( fmw-nu_menos(B_z) )**2+(HWHM)**2) ) - C2*( (HWHM**2)/(( fmw-nu_mas(B_z) )**2+(HWHM)**2) )
```

```python
fmw = np.linspace(2.87-1.3,2.87+1.1,200)
Bz_vals = np.arange(0,20*10**(-3),5*10**(-3)) # En militestas 
plt.figure(figsize=(11,7))

ref = 0 # Esto para que las graficas no se solapen
for B_z in Bz_vals:
    plt.plot(fmw*1000,ref+func(fmw,B_z = B_z),'-b',label=f'B={B_z*10**(3)}mT')
    ref += 0.25
    
plt.xlabel('Frecuencia MW (MHz)')
plt.ylabel('Fluorescencia (u.a.)')
plt.grid(True)
plt.ylim(bottom=0.5) # El limite inferior lo definimos

# Obtener los ejes actuales y eliminar los ticks del eje Y
ax = plt.gca()
ax.set_yticks([])

##### Cargando imagen e introduciendola #############3
img_large = mpimg.imread('diagrama-subniveles-ODMR-unico-centro.jpg')
# Insertar la imagen pequeña dentro de la imagen grande
ax_inset = inset_axes(ax, width="100%", height="100%", bbox_to_anchor=(0.036, -0.06, 0.48, 0.48), 
                      bbox_transform=ax.transAxes)  # 30% del tamaño del eje principal
ax_inset.imshow(img_large)
ax_inset.axis('off')  # Desactivar los ejes para la imagen pequeña

###### Agregamos la diferencia entre dos valles ###########
x1, y1 = nu_menos(10*10**(-3))*10**3 - 10, 1.28  # Valle 1
x2, y2 = nu_mas(10*10**(-3))*10**3 + 10, 1.28 # Valle 2

# Dibujar la línea entre los dos valles
ax.annotate('', xy=(x2, y2), xytext=(x1, y1),
            arrowprops=dict(arrowstyle='<->', color='red',lw=2))

# Calcular el punto medio para el texto
x_mid = (x1 + x2) / 2
y_mid = (y1 + y2) / 2

# Añadir el texto en el punto medio
ax.text(x_mid, y_mid, r'$2\gamma B_{\mathrm{NV}}$', ha='center', va='bottom',fontsize=18)

# Crear un segundo eje Y (derecha)
ax2 = ax.twinx()
ax2.set_yticks([])
ax2.set_ylabel('Incremento de campo B')  # Etiqueta para el eje Y derecho


# Crear una anotación en el gráfico con una flecha hacia arriba
plt.annotate('', xy=(1.04, 1), xycoords='axes fraction', 
            xytext=(1.04, 0), textcoords='axes fraction',
            arrowprops=dict(facecolor='black', arrowstyle='->'),
            horizontalalignment='right', verticalalignment='top')

plt.savefig('ODMR-unico-NV') ###
plt.show()
```
