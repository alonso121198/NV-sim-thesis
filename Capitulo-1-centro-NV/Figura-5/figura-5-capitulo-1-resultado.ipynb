## Matriz del centro NV 
$$ \mathcal{H}/h = D(S_z^{2} - \frac{S(S+1)}{3}) + E(S_x^{2} - S_y^{2}) + \gamma B\cdot S $$

donde:
\begin{align}
D &= 2.87 \; GHz & \gamma &= \frac{g_e \mu_B}{\hbar} = 28 \;GHz/T \\
E &= 5 \; MHz & S&=1
\end{align}

Los operadores de spin 1
\begin{align}
S_x &= \frac{1}{\sqrt{2}}\begin{pmatrix} 0 & 1 & 0 \\ 1 & 0 & 1 \\ 0& 1& 0 \end{pmatrix} &
S_y &= \frac{1}{\sqrt{2}}\begin{pmatrix} 0 & -i & 0 \\ i & 0 & -i \\ 0& i& 0 \end{pmatrix} &
S_z &= \begin{pmatrix} 1 & 0 & 0 \\ 0 & 0 & 0 \\ 0 & 0 & -1 \end{pmatrix}
\end{align}

Se usa los autovectores de $S_z:  \quad \{ |1\rangle = |+1\rangle\;;\; |2\rangle = |0\rangle\;;\; |3\rangle = |-1\rangle \}$



```python
# Escribimos los operadores a usar
from sympy import Matrix, Rational, sqrt, I,latex

# Los operadores espín S=1
S_x = 1/sqrt(2)*Matrix([[0,1,0],[1,0,1],[0,1,0]])
S_y = 1/sqrt(2)*Matrix([[0,-I,0],[I,0,-I],[0,I,0]])
S_z = Matrix([[1,0,0],[0,0,0],[0,0,-1]])
display((S_x,S_y,S_z))
```

```python
# expresamos el Hamiltoniano
from sympy import symbols, eye
D,E,gamma,B_x,B_y,B_z = symbols('D E \gamma B_x B_y B_z', real=True)

H = D*(S_z**2 - Rational(2,3)*eye(3)) + E*(S_x**2 - S_y**2) + gamma*(B_x*S_x + B_y*S_y + B_z*S_z)
print("El hamiltoniano del centro NV es:")
display(H)
```

## Ecuación caracteristica del Hamiltoniano

```python
# Ecuación secular
from sympy import simplify
caracteristica_ecuacion = H.charpoly()
display(caracteristica_ecuacion)
```

## Solución simbolica con $B_x = 0$ y $B_y = 0$ 

```python
### Reemplazando en el hamiltoniano 
Heval = H.subs({B_x:0,B_y:0})
print("Hamiltoniano con B_x = B_y = 0:")
display(Heval)

autovectores = Heval.eigenvects()

print("Autovalor / Autovector")
for i,autovector in enumerate(autovectores):
    print(f"\033[4m{i} autovalor\033[0m")
    display((autovector[0],simplify(autovector[2][0])))
```

## Calculamos las frecuencias $\nu_{-}$ y $\nu_{+}$ 

```python
E_0 = autovectores[0][0]
E_menos = autovectores[1][0]
E_mas = autovectores[2][0]

print("nu_menos")
nu_menos = E_menos - E_0
display(nu_menos)
print(latex(nu_menos))

print("nu_mas")
nu_mas = E_mas - E_0
display(nu_mas)
```

## Grafica bonita para E = 0 y E = 2MHz 

```python
B_rango = np.linspace(-500*10**(-6),500*10**(-6),100) # esta en Tesla
nu_menos = nu_menos.subs({gamma:28,D:2.87})
nu_mas = nu_mas.subs({gamma:28,D:2.87})

# Es necesario definir estas funciones ya que las expresiones anteriores de nu_menos y nu_mas dan valores absolutos
def nu_menos_E0(B_z):
    return 2.87-B_z*28
def nu_mas_E0(B_z):
    return 2.87+B_z*28

################ E = 0 ######################
plt.plot(B_rango*10**(6),nu_menos_E0(B_rango),linestyle='--',color='black',label=r'$\nu_{\pm} = D \pm \gamma B_{z}$')  
plt.plot(B_rango*10**(6),nu_mas_E0(B_rango),linestyle='--',color='black') 
# plt.plot(frecuencias[0],nu_menos,'.k')

################ E = 2 MHz ######################
plt.plot(B_rango*10**(6),[nu_menos.subs({E:2*10**(-3),B_z:B_valor}) for B_valor in B_rango if True],'b',label=r'$\nu_{\pm} = D \pm \sqrt{\gamma^{2} B_{z}^{2}  + E^{2}}$')  
plt.plot(B_rango*10**(6),[nu_mas.subs({E:2*10**(-3),B_z:B_valor}) for B_valor in B_rango if True],'b') 

plt.ylabel('Frecuencia (GHz)')
plt.xlabel(r'Campo magnetico $B_z$ ($\mathrm{\mu}$T)')
# plt.title('Figura 1')
plt.legend()
plt.savefig('figura-frecuencias-Bz') ### 
plt.show()
```
