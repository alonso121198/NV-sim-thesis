```python
# -*- coding: utf-8 -*-
'''
El paquete Imageio dedicada a los datos de imágenes. Ofrece una interfaz sencilla para escribir
y leer una amplia variedad de datos de imágenes: imágenes animads, vídeos, datos volumétricos,
formatos científicos, etc. 

"imread" es una función que se utiliza para leer imágenes de diferentes 
formatos, como JPEG, PNG, BMP, TIFF, entre otros, y cargarlas en un objeto numpy array.
Esta extrayendo una imagen cualquier para conocer dimensiones 
'''
from imageio import imread
'''
El paquete de Python glob proporciona una función para buscar archivos y directorios que coincidan con un patrón especificado. 
La función principal que se utiliza en el paquete glob es glob.glob(), que devuelve una lista de rutas de 
archivos o directorios que coinciden con el patrón especificado.

El patrón utilizado en glob.glob() puede contener caracteres especiales para realizar coincidencias 
más flexibles. Algunos de los caracteres especiales que se pueden usar son:

- *: coincide con cero o más caracteres en un nombre de archivo o directorio.
- ?: coincide con un solo carácter en un nombre de archivo o directorio.
- []: especifica un conjunto de caracteres permitidos para un solo carácter en un nombre de archivo o directorio.
- **: coincide con cualquier cadena de caracteres, incluidos los subdirectorios.
'''
import glob
import numpy as np
import matplotlib.pyplot as plt

#from scipy import misc

#path1 = Path('New folder/')

nmedias = 64
file = list()
for i in range(1,nmedias+1):
    # va a guardar en una lista doble
    file.append(glob.glob(str(i)+'_image_*.png'))

def my_range(start, end, step):
    while start <= end:
        yield start
        start += step


# el rango de frecuencias
x1 = [x for x in my_range(2870-270,2870+270,1)]
      
ODMR = []
for i in range(nmedias):
    n = list()
    for j in range(len(x1)):
        # lee la imagen y crea el array. 
        imag = imread(file[i][j])
        # suma y lo guarda el valor alli.
        n.append(np.sum(imag))
        
    # Al final tiene un array con los valores de intensidad de todas las fotos
    ODMR.append(n)
```

```python
np.array(ODMR).shape
```

#### Graficar algunas muestras

```python
############## Creamos la imagen con la primera, segunda, tercera y cuarta
#plt.figure(figsize = (10,5))
plt.plot(x1,ODMR[0],'-g')
plt.plot(x1,ODMR[1],'-b')
plt.plot(x1,ODMR[2],'-k')
plt.plot(x1,ODMR[3],'-r')
plt.show()
```

#### Promediar

```python
nt = list()
for i in range(len(x1)):
    soma = 0
    for j in range(nmedias):
        soma = soma + ODMR[j][i]
    nt.append(soma)
    
Nnt = list()
norm = float(max(nt))
for i in range(len(x1)):
    Nnt.append(nt[i]/norm)

A = []
for i in range(len(x1)):
	A.append([0]*2)
for i in range(len(x1)):
	A[i][0] = x1[i]
	A[i][1] = Nnt[i]

#np.savetxt('ODMRNOVO.txt',A,fmt='%s')
```

### Graficar el promedio 

```python
plt.figure(figsize = (10,5))
plt.plot(x1,Nnt,'-ob')
plt.tick_params(direction='out', length=4, width=2, colors='black',labelsize=10)
plt.xlabel('Time_MW(MHz)',fontsize=15)
plt.ylabel('Intensity',fontsize=15)
plt.grid()
plt.savefig("../../Graficos_out/ODMR_normal")
#plt.xlim(2765,2770)
#plt.ylim(0.992,1)
plt.show()   
```
